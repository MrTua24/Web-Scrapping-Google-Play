# 1. Pengambilan data ulasan di google play

```python
from google_play_scraper import Sort, reviews_all
import pandas as pd
import numpy as np
```

```python
scrapreview = reviews_all(
    'com.roblox.client',  # ID aplikasi
    lang='id',                 # defaults to 'id'
    country='id',              # defaults to 'id'
    sort=Sort.NEWEST,   # defaults to Sort.MOST_RELEVANT
    count=50,
    filter_score_with=None        # defaults to None (means all score)
)
```

```python
data_roblox = pd.DataFrame(scrapreview)
data_roblox.head()
```

```python
len(data_roblox.index)
```

```python
review_data_roblox = data_roblox[['userName', 'score', 'at', 'content']]
```

```python
review_data_roblox.to_csv("data_review_roblox.csv", index = False)
print ('data berhasil di simpan')
```
